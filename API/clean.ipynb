```python
import pandas as pd
```

```python
df_tel = pd.read_csv(f'email.csv', index_col="Date", parse_dates = True, dayfirst=True)
liste_famille = df_tel.instance.unique()
liste_entite = df_tel.Entite.unique()

df_freq = pd.DataFrame(columns = ['Entite', 'instance','Nb_recus'])

for i, famille in enumerate(liste_famille):
    for j, entite in enumerate(liste_entite):
        df = df_tel[(df_tel["Entite"] == entite)& (df_tel["instance"] == famille)]
        df = df.asfreq('d')
        df.Entite = df.Entite.fillna(entite)
        df.instance = df.instance.fillna(famille)
        df.Nb_recus = df.Nb_recus.fillna(0)
        df_freq = pd.concat([df_freq,df])

df_freq = df_freq.sort_values(['Entite','instance'],ascending = True)
df_freq = df_freq.reset_index(names="Date")

df_freq
df_freq.to_csv('email.csv', index = False)
```
