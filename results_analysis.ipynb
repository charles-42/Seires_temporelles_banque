```python
import pandas as pd
```

```python
df_naive_saisonal = pd.read_csv("data/df_naive_saisonal.csv")
df_naive_drift= pd.read_csv("data/naive_drift.csv")
df_arima = pd.read_csv("data/auto_arima.csv")
df_sarima = pd.read_csv("data/sarima.csv")
df_xgboost = pd.read_csv("data/xgboost.csv")
prophet = pd.read_csv("data/prophet.csv")
```

```python
df = pd.concat([df_naive_saisonal,df_naive_drift,df_arima,df_sarima,df_xgboost,prophet],axis=0).drop(['best_model'],axis=1)
```

```python
liste_famille = df.Famille.unique()
liste_entite = df.Entite.unique()

for i, famille in enumerate(liste_famille):
    for j, entite in enumerate(liste_entite):
        df_filter = df[(df["Entite"] == entite)& (df["Famille"] == famille)]
        
        print(df_filter)
```
